```python
# 18 Find highest revenue month for each year using Window and rank
from pyspark.sql.functions import year, month, sum, col
from pyspark.sql.window import Window
from pyspark.sql.functions import dense_rank
monthly_sales_df = df_sales_order_header \
    .withColumn("order_year", year("OrderDate")) \
    .withColumn("order_month", month("OrderDate")) \
    .groupBy("order_year", "order_month") \
    .agg(sum("TotalDue").alias("monthly_revenue"))
window_spec = Window.partitionBy("order_year").orderBy(col("monthly_revenue").desc())
highest_revenue_month_df = monthly_sales_df.withColumn("month_rank", dense_rank().over(window_spec)) \
    .filter(col("month_rank") == 1)

display(highest_revenue_month_df)

# 19 
from pyspark.sql.functions import to_date
fact_sales_df = df_sales_order_header \
    .join(df_sales_order_detail, "SalesOrderID") \
    .select(
        col("SalesOrderID").alias("sales_order_id"),
        col("CustomerID").alias("customer_key"),
        col("ProductID").alias("product_key"),
        to_date("OrderDate").alias("order_date_key"),
        col("OrderQty"),
        col("UnitPrice"),
        col("UnitPriceDiscount"),
        col("LineTotal"),
        col("TotalDue")
    )

display(fact_sales_df)

# 20 
from pyspark.sql.functions import max, current_timestamp
gold_audit_summary_df = df_sales_order_header.agg(
    max("OrderDate").alias("max_order_date"),
    current_timestamp().alias("refresh_timestamp"),
    countDistinct("SalesOrderID").alias("order_count"),
    countDistinct("CustomerID").alias("customer_count"),
    countDistinct("SalesOrderNumber").alias("unique_order_numbers")
)

display(gold_audit_summary_df)
```
