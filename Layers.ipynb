```python
# 18
from pyspark.sql.functions import rank
monthly_revenue = (
    df_sales_order_header
    .groupBy(
        year("OrderDate").alias("Year"),
        month("OrderDate").alias("Month")
    )
    .agg(
        sum("TotalDue").alias("Revenue")
    )
)
highest_month = (
    monthly_revenue
    .withColumn(
        "Rank",
        rank().over(
            Window.partitionBy("Year")
            .orderBy(desc("Revenue"))
        )
    )
    .filter("Rank=1")
)

highest_month.show()
```
