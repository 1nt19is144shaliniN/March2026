```python
#pow(base, exp[, mod])
#Using pow() with Variables

name = "Shalini"
course = "Data Engineering"

base = 5
exp = 2

result = pow(base, exp)

print(name + " calculated power in " + course)
print("Result:", result)
```

```python
#pow with mod
name = "Shalini"
course = "Data Engineering"

result = pow(2, 3, 5)

print(name + " using modulus in " + course)
print("Result:", result)
```

```python
#divmod(a, b)
name = "Shalini"
course = "Data Engineering"

total_students = 157
students_per_batch = 50

batches, remaining = divmod(total_students, students_per_batch)

print(name + " managing batches in " + course)
print("Total batches:", batches)
print("Remaining students:", remaining)
```
