```python
# Question 1: Calculate the average of 3.14, 2.718, 1.618, 0.577

numbers = [3.14, 2.718, 1.618, 0.577]
avg = sum(numbers) / len(numbers)
print(avg)
```

```python

# Question 3: Calculate the compound interest on $1000 at 5.5% for 3 years

P = 1000
R = 5.5 / 100
T = 3
CI = P * (1 + R) ** T - P
print(CI)
```
