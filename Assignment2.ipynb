```python
# Insert at specific position

numbers = [10, 20, 30, 40]

numbers.insert(1, 15)
numbers.insert(3, 25)

print(numbers)
```

```python
# Remove an item(by value)
numbers = [10, 20, 30, 40]

numbers.remove(20)

print(numbers)
```

```python
#remove by index
numbers = [10, 20, 30, 40]

numbers.pop(2)

print(numbers)
```

```python
#  Find index
numbers = [10, 20, 30, 40]

position = numbers.index(30)

print(position)
```

```python
# count occurence
numbers = [10, 20, 30, 20, 40, 20]

count_20 = numbers.count(20)

print(count_20)
```

```python
# sort in asending order
numbers = [40, 10, 30, 20]

numbers.sort()

print(numbers)
```

```python
# sort in decending order
numbers = [40, 10, 30, 20]

numbers.sort(reverse=True)

print(numbers)
```

```python
# copy the list
numbers = [10, 20, 30, 40]

new_list = numbers[:]

print(new_list)
```

```python
# clear the list
numbers = [10, 20, 30, 40]

numbers.clear()

print(numbers)
```

```python
#  Extend a list
numbers = [1, 2, 3]

numbers.extend([4, 5, 6])

print(numbers)
```
