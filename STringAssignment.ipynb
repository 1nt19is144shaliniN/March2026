```python
# Q41 Balanced parentheses
s = "((()))"
count = 0
balanced = True
for c in s:
    if c == '(':
        count += 1
    else:
        count -= 1
    if count < 0:
        balanced = False
print(balanced and count == 0)
```

```python

# Q43 Longest common substring
s1, s2 = "programming", "grammar"
common = ""
for i in range(len(s1)):
    for j in range(i+1, len(s1)+1):
        if s1[i:j] in s2 and len(s1[i:j]) > len(common):
            common = s1[i:j]
print(common)
```

```python
# Q50 Credit card (Luhn)
def luhn(card):
    digits = [int(d) for d in card]
    for i in range(len(digits)-2, -1, -2):
        digits[i] *= 2
        if digits[i] > 9:
            digits[i] -= 9
    return sum(digits) % 10 == 0

print(luhn("4532015112830366"))
```
